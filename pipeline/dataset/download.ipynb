```python
import requests
import pandas as pd
from itertools import product

api_key = 'MR4TBAKZE9WV75KB'
for y, m in list(product(list(range(2020, 2026)), list(range(1, 13)))):
    month = f'{y}-{m:02d}'
    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=SPY&interval=5min&month={month}&outputsize=full&apikey={api_key}'
    data = requests.get(url).json()
    df = pd.DataFrame(data['Time Series (5min)']).T.iloc[::-1].reset_index()
    df.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
    df.to_csv(f'data/spy_{month.replace("-", "_")}.csv', index=False)
```
